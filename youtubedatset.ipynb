# Bar Graph

```python
y=df["video views"]
x=df["subscribers"]
plt.bar(x,y,label="Youtuber",color="green")
plt.xlabel("subscribers")
plt.ylabel("video views")
plt.title("YOUTUBE VIEWS DATASET")
plt.show()
```

```python
x=df["subscribers"]
plt.hist(x,bins=None,histtype="bar",rwidth=0.5,color='blue')
```

# Pie Chart

```python
slices=[45,24,34,16]
depts=["Rank","Subscribers","Views","Uploads"]
cols=["violet","indigo","green","blue"]
plt.pie(slices,labels=depts,colors=cols,startangle=90,explode=(0,0,0,0.2),shadow=True,autopct="%.1f%%")
```
